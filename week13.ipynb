```python
import csv
import matplotlib.pyplot as plt
f=open('seoul.csv')
data=csv.reader(f)
next(data)
result=[]
min_result=[]
for row in data:
    if row[1] !='':
        if row[0].split('-')[1]=='11' and row[0].split('-')[2]=='26':
            result.append(float(row[-1]))
            min_result.append(float(row[-2]))
plt.plot(result,label='max_temp')
plt.plot(min_result,label='min_temp')
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
aug=[]
jan=[]
f=open('seoul.csv')
data=csv.reader(f)
next(data)
for row in data:
    month=row[0].split('-')[1]
    
    if row[-1] !='':
        if month=='08':
            aug.append(float(row[-1]))
        elif month=='01':
            jan.append(float(row[-1]))

plt.hist(aug,bins=100,color='b',label='Aug')
plt.hist(jan,bins=100,color='r',label='Jan')
plt.legend()
plt.show()
```
