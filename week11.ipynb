```python
class Calc:
    def sum(self,a,b):
        result=a+b
        print("{0:s}+{1:s}={2:s} 입니다.".format(str(a),str(b),str(result)))

    def sub(self,a,b):
        result=a-b
        print("{0:s}-{1:s}={2:s} 입니다.".format(str(a),str(b),str(result)))
    def multi(self,a,b):
        result=a*b
        print("{0:s}*{1:s}={2:s} 입니다.".format(str(a),str(b),str(result)))
    def divi(self,a,b):
        result=a/b
        print("{0:s}/{1:s}={2:s} 입니다.".format(str(a),str(b),str(result)))

calc = Calc()
calc.sum(1,2)
calc.sub(5,1)
calc.multi(2,3)
calc.divi(6,2)
```

```python
class ArrayStack:
    
    def __init__ (self):
        self.top = []
    
    def size(self):
        return len(self.top)

    def push (self, item):
        self.top.append(item)
    def pop(self):
        if not self.isEmpty():
            return self.top.pop(-1)
        else:
            print("Stack underflow")
            exit()
    
    def peek(self):
        if not self.isEmpty():
            return self.top[-1]
        else:
            print("Stack underflow")
            exit()

    def isEmpty(self):
        return len(self.top)==0
    
    def size(self):
        return len(self.top)

userinput=input()
input_list=userinput.split(" ")
for char in input_list:
    print(char,end="")
print("=",end="")
def split(Str):
    List = []
    val = 0                 
    checking = False   
    
    for char in Str:
        if char == ' ':
            continue
        if char in '0123456789' :
            val = val * 10 + int(char)
            checking = True
        else:
            if checking:
                List.append(val)
                val = 0
            
            checking = False
            List.append(char)
        
    if checking:
        List.append(val)
        
    return List

def infixToPostfix(List):
    prec = {
        '*' : 3,
        '/' : 3,
        '+' : 2,
        '-' : 2,
        '(' : 1
    }
    
    opStack = ArrayStack()
    postfixList = []
    
    for item in List:
        if type(item) is int:
            postfixList.append(item)
            
        elif item == '(':
            opStack.push(item)

        elif item == ')':
            while opStack.peek() != '(':
                postfixList.append(opStack.pop())
            opStack.pop()
        
        else:
            if opStack.isEmpty():
                opStack.push(item)
                
            else:
                while opStack.size() > 0:
                    if prec[opStack.peek()] >= prec[item]:
                        postfixList.append(opStack.pop())
                    else:
                        break
                
                opStack.push(item)
    
    while not opStack.isEmpty():
        postfixList.append(opStack.pop())
    

    return postfixList


def calc(List):
    Stack = ArrayStack()
    
    for item in List:
      
        if type(item) is int:
            Stack.push(item)
            
 
        elif item == '+':
            n1 = Stack.pop()
            n2 = Stack.pop()
            Stack.push(n2+n1)
        
        elif item == '-':
            n1 = Stack.pop()
            n2 = Stack.pop()
            Stack.push(n2-n1)
            
        elif item == '*':
            n1 = Stack.pop()
            n2 = Stack.pop()
            Stack.push(n2*n1)
            
        elif item == '/':
            n1 = Stack.pop()
            n2 = Stack.pop()
            Stack.push(int(n2/n1))
    return Stack.pop()
print(calc(infixToPostfix(split(input_list))))
```
