```python
import csv


f=open('subwayuse.csv',encoding='cp949')

data=csv.reader(f)
data_list=[]
next(data)
max_list=[0,0,0,0]
max_name=[""]*4
min_list=[9999999,9999999,9999999,9999999]
min_name=[""]*4
div_station=[]
temp=[]
for row in data:
    data_list.append(row)
    
for i in range(0,len(data_list)):
  
    try:
        if(data_list[i][1]==data_list[i+1][1]):
            temp.append(data_list[i])
        else:
           
            div_station.append(temp)
            temp=[]
    except:
       
        pass

for i in range(0,len(div_station)):
    for row in div_station[i]:   
        row[-2]=row[-2].replace(',','')
        row[-2]=int(row[-2])
        row[-1]=row[-1].replace(',','')
        row[-1]=int(row[-1])

        using_num=row[-2]+row[-1]
        max_name[0]=row[3]+' '+row[1]
        max_list[0]=using_num
        min_name[0]=row[3]+' '+row[1]
        min_list[0]=using_num
        for i in range(1,len(max_list)):
            for j in range(i,0,-1):
                if max_list[j] < max_list[j-1]:
                    max_list[j], max_list[j-1] = max_list[j-1], max_list[j]
                    max_name[j], max_name[j-1] = max_name[j-1], max_name[j]
                else :
                    break

        for i in range(1,len(min_list)):
            for j in range(i,0,-1):
                if min_list[j] > min_list[j-1]:
                    min_list[j], min_list[j-1] = min_list[j-1], min_list[j]
                    min_name[j], min_name[j-1] = min_name[j-1], min_name[j]
                else :
                    break
    print("이용객이 가장 많은 역 순위")
    for i in range(3,0,-1):
        j=0
        if i==3:
            j=1
        elif i==2:
            j=2
        elif i==1:
            j=3
        print(str(j)+". "+max_name[i]+"("+str(max_list[i])+"명)")
    print("")
    print("이용객이 가장 적은 역 순위")
    for i in range(3,0,-1):
        
        j=0
        if i==3:
            j=1
        elif i==2:
            j=2
        elif i==1:
            j=3
        print(str(j)+". "+min_name[i]+"("+str(min_list[i])+"명)")
    print("")
    max_list=[0,0,0,0]
    max_name=[""]*4
    min_list=[9999999,9999999,9999999,9999999]
    min_name=[""]*4
    


f.close()
```
